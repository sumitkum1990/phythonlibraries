```python
#3d plot

x = np.random.rand(20)
y = np.random.rand(20)
z = np.random.rand(20)

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x,y,z,c = "#FF5733", alpha = .9)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
#3d plot

x = np.random.rand(20)
y = np.random.rand(20)
z = np.random.rand(20)

fig = plt.figure()
ax = fig.add_subplot(projection = '4d')
ax.scatter(x,y,z,c = "#FF5733", alpha = .9)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```
